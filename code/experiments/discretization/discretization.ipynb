```python
# run this after the script to generate plots
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

result = pd.read_csv("result.csv")
```

```python
result.columns = ["0","n_discrete","price_score", "carbon_score"]
```

```python
result
```

```python
fig, ax = plt.subplots()
ax.plot(result["n_discrete"], result["carbon_score"], label="Carbon Cost", marker="o")
ax.plot(result["n_discrete"], 0.5*(result["price_score"]+ result["carbon_score"]), label="Average Cost", marker=".")
ax.plot(result["n_discrete"], result["price_score"], label="Dollar Cost", marker="x")
ax.add_patch(plt.Rectangle((8.5,1),1,-0.25, edgecolor="black", fill=False, lw=1.5))
ax.legend()
ax.set_xticks([0,2,4,6,8,9,10, 12, 14, 16, 18, 20])
plt.ylabel("Normalized Cost")
plt.xlabel("Possible Discrete Actions")
plt.savefig("discretization.png")
plt.show()
```
